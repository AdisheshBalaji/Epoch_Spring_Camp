```python
class K_Means:
    def __init__(self, n_clusters, tol = 0.0001, max_iter = 10):
        self.n_clusters = n_clusters
        self.tol = tol 
        self.max_iter = max_iter 

    def fit(self, data):
        self.centroids = {}

        for i in range(self.n_clusters):
            self.centroids[i] = data[i]

        for _ in range(self.max_iter):
            self.classifications = {}

            for i in range(self.n_clusters):
                self.classifications[i] = []
            
            for features in data:
                distances = []
                for centroid in self.centroids:
                    distance = np.linalg.norm(features - self.centroids[centroid])
                    distances.append(distance)
                
                classification = distances.index(min(distances))
                self.classifications[classification].append(features)

            prev_centroids = dict(self.centroids)

            for classification in self.classifications:
                self.centroids[classification] = np.average(self.classifications[classification], axis = 0)

            converged = True

            for c in self.centroids:
                original_centroid = prev_centroids[c]
                current_centroid = self.centroids[c]

                if np.linalg.norm(current_centroid - original_centroid) > self.tol:
                    converged = False

            if converged:
                break





    def predict(self, data):
        distances = []
        for centroid in self.centroids:
            distance = np.linalg.norm(centroid - data)
            distances.append(distance)
        classification = distances.index(min(distances))
        return classification
    
    
        
```
