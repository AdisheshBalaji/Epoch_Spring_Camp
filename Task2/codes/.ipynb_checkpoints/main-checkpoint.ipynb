```python
northern_limit = [19.888774, 78.321238]
southern_limit = [16.030649, 78.152414]
eastern_limit = [17.897977, 80.820099]
western_limit = [18.030339, 77.746985]
```

```python
max_latitude = northern_limit[0]
min_latitude = southern_limit[0]
max_longitude = eastern_limit[1]
min_longitude = western_limit[1]
```

```python
tg_df = telangana_df[telangana_df["Latitude"].between(min_latitude, max_latitude) &
                         telangana_df["Longitude"].between(min_longitude, max_longitude)]
```

```python
latitudes = tg_df["Latitude"].to_numpy()
longitudes = tg_df["Longitude"].to_numpy()
```

```python
plt.scatter(latitudes, longitudes, s = 1)
```

```python
class K_Means:
    def __init__(self, n_clusters, tol = 0.0001, max_iter = 10):
        self.n_clusters = n_clusters
        self.tol = tol 
        self.max_iter = max_iter 

    def fit(self, data):
        self.centroids = {}

        for i in range(self.n_clusters):
            self.centroids[i] = data[i]

        for _ in range(self.max_iter):
            self.classifications = {}

            for i in range(self.n_clusters):
                self.classifications[i] = []
            
            for features in data:
                distances = []
                for centroid in self.centroids:
                    distance = np.linalg.norm(features - self.centroids[centroid])
                    distances.append(distance)
                
                    classification = distances.index(min(distances))
                    self.classifications[classification].append(features)

            prev_centroids = dict(self.centroids)

            for classification in self.classifications:
                self.centroids[classification] = np.average(self.classifications[classification], axis = 0)

            converged = True

            for c in self.centroids:
                original_centroid = prev_centroids[c]
                current_centroid = self.centroids[c]

                if np.sum((current_centroid - original_centroid)/original_centroid * 100) > self.tol:
                    converged = False

            if converged:
                break





    def predict(self, data):
        distances = []
        for centroid in self.centroid:
            distance = np.linalg.norm(centroid - data)
            distances.append(distance)
        classification = distances.index(min(distances))
        return classification
    
    
        
```

```python
model=K_Means(5)
colors = ["b", "o", "r", "g", "y", "v"]
data = np.column_stack((latitudes, longitudes))
model.fit(data)
print(model.centroids)

```

```python
import matplotlib.pyplot as plt

colors = ["b", "r", "g", "y", "c", "m"]  # one color per cluster

# Plot each cluster
for cluster_id in model.classifications:
    cluster_points = model.classifications[cluster_id]
    color = colors[cluster_id % len(colors)]
    
    # Plot points belonging to this cluster
    for point in cluster_points:
        plt.scatter(point[0], point[1], s=1, color=color)
    
    # Plot the centroid
    centroid = model.centroids[cluster_id]
    plt.scatter(centroid[0], centroid[1], color=color, marker='X', s=10, edgecolors='k', linewidths=2)

plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("K-Means Clustering with Colored Centroids")
plt.show()
```
